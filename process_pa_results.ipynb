```python
import csv

def process_oed(file, year):
    county_results = {}
    with open(file,'r') as f:
        reader = csv.DictReader(f)
        print(reader.fieldnames)
        c = True
        for row in reader:
            if c:
                print(row)
                c = False
            if row['office'] != "President":
                continue
            county = row['parent_jurisdiction'].capitalize()
            candidate = (row['first_name'].capitalize() + " " + row['last_name'].capitalize()).strip()
            votes = int(row['votes'])
            county_results.setdefault(county,{}).setdefault(candidate,[]).append(votes)

    results = []
    for county, r in county_results.items():
        c_r = {}
        c_r["County"] = county
        for candidate, votes in r.items():
            c_r[candidate] = sum(votes)
        results.append(c_r)

    print(county_results["Adams"])
    print(results[0])

    with open('PA/'+str(year)+'_county.csv','w') as csvfile:
        fields = list(results[0].keys())
        fields.remove("County")
        fields = ['County'] + sorted(fields)
        writer = csv.DictWriter(csvfile, fieldnames=fields, restval=0, quoting=csv.QUOTE_NONNUMERIC)

        writer.writeheader()
        for r in results:
            writer.writerow(r)

process_oed('PA/20001107__pa__general__precinct__raw.csv', 2000)
process_oed('PA/20041102__pa__general__precinct__raw.csv', 2004)
process_oed('PA/20081104__pa__general__precinct__raw.csv', 2008)
process_oed('PA/20121106__pa__general__precinct__raw.csv', 2012)
```
